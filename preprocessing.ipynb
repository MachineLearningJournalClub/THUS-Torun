# Preprocessing

## Smaller units

```python
import cltk
from cltk import NLP
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text
```

```python
raw=clean_text(raw)
```

```python
cltk_nlp = NLP(language="lat")
```

```python
cltk_doc = cltk_nlp.analyze(text=raw)
```
